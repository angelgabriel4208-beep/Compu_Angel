```python
import pandas as pd
import numpy as np

 
nombres = ['Ana', np.nan, 'Carlos', 'Marta', 'Sofia', 'Diego', np.nan, 'Pedro', 'Lucia', 'Jorge']
edades = [20, 30, 23, 43, 23, 65, np.nan, 34, 43, 56]
salarios = [2300, 4320, 4323, np.nan, 7665, 2343, 2356, 4567, 7656, 3000]
puestos = [np.nan, 'Ingeniero', 'Vendedor', 'Contador', 'Abogado', 'Soldador', 'Empresario', 'Ingeniero', 'Tecnico', np.nan]

df_trabajadores = pd.DataFrame({
    'Nombre': nombres,
    'Edad': edades,
    'Puesto': puestos,
    'Salario': salarios
})

df_trabajadores

dfNulos = df_trabajadores.isnull()

print(df_trabajadores)
print(dfNulos)


dfSum = df_trabajadores.isnull().sum()

print(dfSum)

dfCorrectos = df_trabajadores.notnull()

print(dfCorrectos)
```
